```python
# Import libaries
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
import cv2
import os
import glob
```

```python
# img Testing for Har Cascade libary
img = cv2.imread("images_dataset\lionel_messi\_111066400_messi.jpg")
plt.imshow(img)
```

```python
# Importing HarCascade Detection files from OpenCV

# Face CasCade file
face_cascade = cv2.CascadeClassifier("opencv\haarcascades\haarcascade_frontalface_default.xml")

# Eye CasCade file
eye_cascade = cv2.CascadeClassifier("opencv\haarcascades\haarcascade_eye.xml")

# Printing the array of Face Cascade value
face_arr = face_cascade.detectMultiScale(img,  1.3, 5)
face_arr # returns 4 val [x, y, w, h]
```

```python
# Assuming you already have the image 'img' and 'face_arr' with the detected face coordinates

for (x, y, w, h) in face_arr:               # Color: Blue (255, 0, 0)
    cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2) # 2 is the thickness of the border of Rectangle

# The above code draws a blue rectangle around each detected face in 'img'

# Finally, you can display the image with the rectangles using cv2.imshow or save it using cv2.imwrite.
cv2.imshow('Detected Faces', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Detect eyes from image
eye_arr = eye_cascade.detectMultiScale(img, 1.3, 5)
eye_arr
```

```python
# Present those eyes in Circle Red
for (x, y, w, h) in eye_arr:
    # Calculate the center of the eye
    center_x = x + w // 2
    center_y = y + h // 2
    
    # Calculate the radius of the circle (approximately half of the width or height)
    radius = max(w, h) // 2
    
    # Draw a circle on the image
    cv2.circle(img, (center_x, center_y), radius, (0, 0, 255), 2)

# Display the image with circles around the detected eyes
cv2.imshow('Eyes Detected', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Print out the ROI's [ Region of Intrest]

# Create a list to store the face ROIs
face_rois = []

# Loop through detected faces
for (x, y, w, h) in face_arr:
    # Define the region of interest (ROI) for the face
    roi_face = img[y:y + h, x:x + w]
    face_rois.append(roi_face)

# Plot the face ROIs using Matplotlib
fig = plt.figure(figsize=(12, 4))

for i, roi in enumerate(face_rois):
    plt.subplot(1, len(face_rois), i + 1)
    plt.imshow(cv2.cvtColor(roi, cv2.COLOR_BGR2RGB))
    plt.axis('off')  # Turn off the axis labels

plt.show()
```

```python
def croppedImages(image_path):
    # Read the image from the specified file path
    img = cv2.imread(image_path)
    
    # Convert the loaded image to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # Detect faces in the grayscale image using the face_cascade classifier
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5)
    
    # Loop through the detected faces
    for (x, y, w, h) in faces:
        # Extract the grayscale region of interest (ROI) for the detected face
        roi_gray = gray[y:y+h, x:x+w]
        
        # Extract the color ROI for the detected face
        roi_color = img[y:y+h, x:x+w]
        
        # Detect eyes within the grayscale ROI
        eyes = eye_cascade.detectMultiScale(roi_gray)
        
        # Check if at least 2 eyes are detected within the ROI
        if len(eyes) >= 2:
            # If 2 or more eyes are detected, return the color ROI (face with eyes)
            return roi_color
    
    # If no face with 2 or more eyes is found, return None
    return None
```

```python
def croppedImages(image_path):
    # Read the image from the specified file path
    img = cv2.imread(image_path)
    
    # Convert the loaded image to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # Detect faces in the grayscale image using the face_cascade classifier
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5)
    
    # Loop through the detected faces
    for (x, y, w, h) in faces:
        # Extract the grayscale region of interest (ROI) for the detected face
        roi_gray = gray[y:y+h, x:x+w]
        
        # Extract the color ROI for the detected face
        roi_color = img[y:y+h, x:x+w]
        
        # Detect eyes within the grayscale ROI
        eyes = eye_cascade.detectMultiScale(roi_gray)
        
        # Check if at least 2 eyes are detected within the ROI
        if len(eyes) >= 2:
            # If 2 or more eyes are detected, return the color ROI (face with eyes)
            return roi_color
    
    # If no face with 2 or more eyes is found, return None
    return None

# Define the path where cropped images will be saved
cropDir_path = 'D:\Git\ImageClassifier_Athletics\images_dataset\Cropped'

# Initialize empty lists and a dictionary to store information
cropped_img_dir = []  # Empty list to store cropped images
celeb_file_name_dict = {}  # Empty dictionary to store celebrity names and their corresponding cropped images

# Loop through each subdirectory in the img_dirs list
for img_dir in img_dirs:
    count = 1
    # Extract the celebrity name from the directory path
    celebrity_name = os.path.basename(img_dir)  # Get the name of the celebrity from the directory
    
    # Loop through each file in the subdirectory
    for entry in os.scandir(img_dir):
        # Call the 'croppedImages' function to process the image
        roi_color = croppedImages(entry.path)
        
        # Check if 'roi_color' is not None (i.e., a valid cropped image)
        if roi_color is not None:
            # Define the path for the celebrity's specific cropped folder
            cropped_folder = os.path.join(cropDir_path, celebrity_name)
            
            # Create the cropped folder if it doesn't exist
            if not os.path.exists(cropped_folder):
                os.makedirs(cropped_folder)
                
                # Append the path of the generated cropped folder to the list
                cropped_img_dir.append(cropped_folder)
                print(f"Cropped folders generated successfully for {cropped_folder}")
                
            # Generate a unique cropped file name for each image
            cropped_file_name = celebrity_name + str(count) + '.png'
            cropped_file_path = os.path.join(cropped_folder, cropped_file_name)
            
            # Save the cropped image to the appropriate folder
            cv2.imwrite(cropped_file_path, roi_color)
            
            # Add the cropped file path to the celebrity's list in the dictionary
            if celebrity_name not in celeb_file_name_dict:
                celeb_file_name_dict[celebrity_name] = []
            
            celeb_file_name_dict[celebrity_name].append(cropped_file_path)
            
            count += 1
```

```python
import os
import cv2
import glob

# Initialize paths
data_path = 'D:\Git\ImageClassifier_Athletics\images_dataset'
cropped_path = os.path.join(data_path, 'Cropped')

# Initialize the face and eye cascade classifiers
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')

# Create the 'Cropped' directory if it doesn't exist
os.makedirs(cropped_path, exist_ok=True)

# Function to crop images if a face with at least 2 eyes is detected
def crop_and_save_images(img_files):
    for img_file in img_files:
        # Read the image file
        img = cv2.imread(img_file)
        
        # Convert the image to grayscale
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        
        # Detect faces in the grayscale image
        faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5)

        for i, (x, y, w, h) in enumerate(faces):
            # Extract the region of interest (ROI) for the detected face
            roi_gray = gray[y:y+h, x:x+w]
            
            # Detect eyes within the grayscale ROI
            eyes = eye_cascade.detectMultiScale(roi_gray)
            
            # Check if at least 2 eyes are detected
            if len(eyes) >= 2:
                # Get the celebrity name from the image file path
                celeb_name = os.path.basename(os.path.dirname(img_file))
                
                # Create the celebrity's cropped folder if it doesn't exist
                celeb_cropped_folder = os.path.join(cropped_path, celeb_name)
                os.makedirs(celeb_cropped_folder, exist_ok=True)
                
                # Generate a unique cropped file name for each image
                cropped_file_name = f"{celeb_name}_{i+1}.png"
                cropped_file_path = os.path.join(celeb_cropped_folder, cropped_file_name)
                
                # Save the cropped image to the appropriate folder
                cv2.imwrite(cropped_file_path, img[y:y+h, x:x+w])

# Use glob to collect image files
img_files = []
for img_type in ['*.jpg', '*.jpeg', '*.png', '*.bmp', '*.gif', '*.jfif', '*.webp']:
    img_files.extend(glob.glob(os.path.join(data_path, '**', img_type), recursive=True))

# Process and save images
crop_and_save_images(img_files)
```
