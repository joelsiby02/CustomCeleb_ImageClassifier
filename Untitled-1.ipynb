```python
# Detect eyes from image
eye_arr = eye_cascade.detectMultiScale(img, 1.3, 5)
eye_arr
```

```python
# Present those eyes in Circle Red
for (x, y, w, h) in eye_arr:
    # Calculate the center of the eye
    center_x = x + w // 2
    center_y = y + h // 2
    
    # Calculate the radius of the circle (approximately half of the width or height)
    radius = max(w, h) // 2
    
    # Draw a circle on the image
    cv2.circle(img, (center_x, center_y), radius, (0, 0, 255), 2)

# Display the image with circles around the detected eyes
cv2.imshow('Eyes Detected', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Print out the ROI's [ Region of Intrest]

# Create a list to store the face ROIs
face_rois = []

# Loop through detected faces
for (x, y, w, h) in face_arr:
    # Define the region of interest (ROI) for the face
    roi_face = img[y:y + h, x:x + w]
    face_rois.append(roi_face)

# Plot the face ROIs using Matplotlib
fig = plt.figure(figsize=(12, 4))

for i, roi in enumerate(face_rois):
    plt.subplot(1, len(face_rois), i + 1)
    plt.imshow(cv2.cvtColor(roi, cv2.COLOR_BGR2RGB))
    plt.axis('off')  # Turn off the axis labels

plt.show()
```

```python
def croppedImages(image_path):
    try:
        # Attempt to read the image from the specified file path
        img = cv2.imread(image_path)
        
        # Check if the image was loaded successfully
        if img is None:
            print(f"Failed to load image: {image_path}")
            return None
        
        # Convert the loaded image to grayscale
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5)
        
        # Loop to Crop the Region of Interest (ROI) and detect eyes
        for (x, y, w, h) in faces:
            # Extract the grayscale ROI from the original image
            roi_gray = gray[y: y+h, x: x+w]  # Grayscale
            # Extract the color ROI from the original image
            roi_color = img[y: y+h, x: x+w]  # Color

            # Detect eyes within the grayscale ROI
            eyes = eye_cascade.detectMultiScale(roi_gray)

            # Check if at least 2 eyes are detected within the ROI
            if len(eyes) >= 2:
                # If at least 2 eyes are detected, return the color ROI (face with eyes)
                return roi_color

        # If no valid face with 2 or more eyes was found, return None
        return None
    except Exception as e:
        print(f"An error occurred: {str(e)}")
        return None
```

```python
import numpy as np
import pywt
import cv2    

def w2d(img, mode='haar', level=1):
    imArray = img
    #Datatype conversions
    #convert to grayscale
    imArray = cv2.cvtColor( imArray,cv2.COLOR_RGB2GRAY )
    #convert to float
    imArray =  np.float32(imArray)   
    imArray /= 255;
    # compute coefficients 
    coeffs=pywt.wavedec2(imArray, mode, level=level)

    #Process Coefficients
    coeffs_H=list(coeffs)  
    coeffs_H[0] *= 0;  

    # reconstruction
    imArray_H=pywt.waverec2(coeffs_H, mode);
    imArray_H *= 255;
    imArray_H =  np.uint8(imArray_H)

    return imArray_H
```

```python
class_dict = {}
count = 0

for celebrity_name in celeb_file_name_dict.keys():
    class_dict[celebrity_name] = count 
    count +=  1
class_dict
```

```python
# Initialize empty lists to store features (X) and labels (y)
X, y = [], []

# Iterate through each celebrity's training files
for celebrity_name, training_files in celeb_file_name_dict.items():
    for training_image in training_files:
        # Read the image file using OpenCV
        img = cv2.imread(training_image)
        
        # Check if the image loading failed
        if img is None:
            continue
        
        # Resize the raw image to 32x32 pixels
        scalled_raw_img = cv2.resize(img, (32, 32))
        
        # Apply a wavelet transform (e.g., 'db1' wavelet) with a specified level (e.g., level=5)
        img_har = w2d(img, 'db1', 5)
        
        # Resize the wavelet transformed image to 32x32 pixels
        scalled_img_har = cv2.resize(img_har, (32, 32))
        
        # Stack the flattened pixel values of the scaled raw image and scaled wavelet transformed image vertically
        combined_img = np.vstack((scalled_raw_img.reshape(32 * 32 * 3, 1), scalled_img_har.reshape(32 * 32, 1)))
        
        # Append the combined image and its corresponding label to X and y, respectively
        X.append(combined_img)
        y.append(class_dict[celebrity_name])
```

```python
# NumPy arrays are often used in numerical computations and provide efficient data storage and manipulation.

X = np.array(X).reshape(len(X), 4096).astype(float) # Reshaping & converting int to float
```

```python
# Import necessary libraries of sklearn
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.pipeline import Pipeline
```

```python
# Splitting into Train & Test Dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
```

```python
# Create a base pipeline with feature scaling and an SVC classifier with specific hyperparameters
pipe = Pipeline([
    ('scaler', StandardScaler()),  # Standardize the features
    ('svc', SVC(kernel='rbf', C=10))  # Support Vector Classifier with 'rbf' kernel and C=10
])

# We will find the best paramter later below to fix our best params model 
```

```python
# Fit the pipeline to the training data
pipe.fit(X_train, y_train)

# Evaluate the model's accuracy on the test data
accuracy = pipe.score(X_test, y_test)

# Print the accuracy score
print("Accuracy:", accuracy)
```
